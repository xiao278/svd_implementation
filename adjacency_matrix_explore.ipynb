```python
import numpy as np
import rustworkx as rx
from rustworkx.visualization import graphviz_draw
from basic_SVD import find_SVD, recombine_SVD
```

```python
# (node-to, edge strength)
directed_adjacency_list = [
    [(1,1),(3,2),(5,1)],
    [],
    [(1,1),(5,2),(3,1)],
    [],
    [(5,1),(1,2),(3,1)],
    []
]

n = len(directed_adjacency_list)
```

```python
adjacency_matrix = np.zeros((n,n))

# fill adj mat from adj list
for root_node in range(0, n):
    for edge in directed_adjacency_list[root_node]:
        to_node, value = edge
        adjacency_matrix[root_node, to_node] = value

print(adjacency_matrix.dtype)

graph = rx.PyDiGraph.from_adjacency_matrix(adjacency_matrix)
graphviz_draw(graph)
```

```python
U, S, V = find_SVD(adjacency_matrix)

first_layer_recombine = recombine_SVD(U, S, V, 1)
new_graph = rx.PyDiGraph.from_adjacency_matrix(first_layer_recombine)
graphviz_draw(new_graph)
```
